```python
Y = df[['G' , 'A' , 'B', 'C']]
Y
```

```python
for col in Y.columns:
    print(f"Unique values in column {col}: {Y[col].unique()}")
```

```python
df.drop('G', axis=1, inplace=True)
df.drop('A', axis=1, inplace=True)
df.drop('B', axis=1, inplace=True)
df.drop('C', axis=1, inplace=True)
df
```

```python
X = df
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=300, criterion='gini', random_state=0)
classifier.fit(X_train,Y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import multilabel_confusion_matrix
from sklearn.metrics import classification_report, accuracy_score
```

```python


# Calculate confusion matrix for each label
cm = multilabel_confusion_matrix(Y_test, y_pred)

# Print confusion matrices
for i, matrix in enumerate(cm):
    print(f"Confusion matrix for label {i}:")
    print(matrix)
```

```python
accuracy_score(Y_test, y_pred)
```

```python
print(classification_report(Y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.plot(Y_test, label='Actual')
plt.plot(y_pred, label= 'Predicted')
plt.xlabel('Sample')
plt.ylabel('Fault')
plt.title('Actual vs Predicted')
plt.legend()
plt.show()
```
